## reshaping, stacking, squeezing and unsqueezing tensors

* reshaping - reshapes tensor to some defined shape
* view - return a view of an input tensor of a certain shape but keep the same memory as the original tensor (shares the same memory as the original tensor)
* stacking - combine multiple tensors. on top of each other (vstack) or side by side (hstack)
* squeeze - removes all `1` dimensions from a target tensor
* unsqueeze - adds a `1` dimension to a target tensor
* premute - return a view of input with dimensions permuted (swapped) in a certain way.

```python
import torch

x = torch.arange(1, 10);
x.shape
```

```python
# reshaping tensor - changing the shape
x_reshaped = x.reshape(1,9); # reshaped into
x_reshaped
```

```python
# changing the view
z = x.view(1,9)
z[:,0] = 5
z,x # change in x chnages y because view of the tensor shares the same memory as original input
```

```python
torch.stack((x,x,x,x),dim=1)

m = torch.ones(2,3);
n = torch.zeros(2,3);

torch.stack((m,n),0)
```

```python
# torch.squeeze() - removes all the single dimensions
torch.squeeze(torch.randn(2,3,1,4,1,1)).shape
```

```python
# toch.unsqeeze - adds a single dimension to a target tensor at specific dim
torch.unsqueeze(torch.randn(2,3,1,4),2)
```

```python
# rearranges the dimensions of target tensor in a specified order
x_original = torch.randn(224,224,3) # [height, width, color channel]

# permute the original tensor to chnage the order od axis (dims)
x_permuted = x_original.permute((2,0,1))
print(f"orginal shape {x_original.shape}")
print(f"permuted shape {x_permuted.shape}")

# the permuted value is a view of original tensor so they share same memory
x_permuted[:, :, 0] = 33333.00
print(f"orginal {x_original}")
print(f"permuted {x_permuted}")
```

### Since NumPy is a popular Python numerical computing library, PyTorch has functionality to interact with it nicely.

The two main methods you'll want to use for NumPy to PyTorch (and back again) are:

torch.from_numpy(ndarray) - NumPy array -> PyTorch tensor.
torch.Tensor.numpy() - PyTorch tensor -> NumPy array.

```python
import torch
# import numpy as np

npArray = np.arange(1.0,9.0)
tensor = torch.from_numpy(npArray) # converting from numpy -> pytorch tensor, pytorch reflect the numpy default float64 dtpe unless specified
```

```python
# change the value of tensor, what this will do to tensor?
tensor[0] = 444.0
npArray, tensor
```

```python
# tensor to numpy array
tensor = torch.ones((7))
tensor_numpy = tensor.numpy()
tensor_numpy, tensor
```

```python
tensor = tensor + 1
tensor[0] =444.0
tensor_numpy, tensor
```

### Reproducibility (trying to take random out of random)

The way a neural network learns:
` start with random number -> tensor operations -> update random numbers to better represnt the data -> again -> again ...`

to reduce randomness in pytorch and neural networks, we use the concept of **random seed**

Essentailly what rando seed does is "flavour" the randomness

```python
random_A = torch.randn(2,3)
random_B = torch.randn(2,3)

print(random_A == random_B)
```

```python
import torch

RANDOM_SEED = 42

torch.manual_seed(RANDOM_SEED)
random_C = torch.randn(3,2)

torch.manual_seed(RANDOM_SEED)
random_D = torch.randn(3,2)

random_C , random_D, random_C == random_D
```
